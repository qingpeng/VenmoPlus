```python
import requests

import json
def hit_api():
    page = "https://venmo.com/api/v5/public"
    venmos = []
    counter = 0
    while (counter < 50):
        r = requests.get(page)
        page = r.json["paging"]["next"]
        venmos += r.json["data"]
        counter += 1
    return venmos
```
